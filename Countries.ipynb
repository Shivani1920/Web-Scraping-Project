```python
from bs4 import BeautifulSoup
import requests
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area#References'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find('table',class_="wikitable sortable sticky-header col2left")
column = table.find_all('th')
table_column = [columns.text.strip() for columns in column]
```

```python
import pandas as pd
df = pd.DataFrame([table_column])
```

```python
all_rows = []
rows = table.find_all('tr')
for row in rows[1:]:
    data = row.find_all('td')
    row_data = [datas.text.strip() for datas in data]
    all_rows.append(row_data)
```

```python
length = len(df)
df.loc[length] = row_data
df = pd.DataFrame(all_rows, columns=table_column)
df
```
